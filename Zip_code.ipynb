```python
from math import radians, sin, cos, sqrt, atan2
def haversine(lat1, lon1, lat2, lon2):
    R= 6371 #earth radius in km
    lat1, lon1, lat2, lon2 = map(radians, [lat1,lon1,lat2,lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2*atan2(sqrt(a),sqrt(1-a))
    return R*c
def zip_distance(zip1, zip2):
  if zip1 in zip_code and zip2 in zip_code:
     lat1, lon1 = zip_code[zip1]['latitude'], zip_code[zip1]["longitude"]
     lat2, lon2 = zip_code[zip2]["latitude"], zip_code[zip2]['longitude']
     return haversine(lat1, lon1, lat2, lon2)
  else:
     return "Zip code not found"
zip1, zip2 = '2808', '2909'
distance_km = zip_distance(zip1, zip2)
print(distance_km)
```
